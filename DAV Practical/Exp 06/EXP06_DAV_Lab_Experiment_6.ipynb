```python
# import libraries
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
```

```python
df = pd.read_csv('amazon.csv')

df
```

```python
# Create preprocess_text function
def preprocess_text(text):
    # Tokenize the text
    tokens = word_tokenize(text.lower())
    # Remove stop words
    filtered_tokens = [token for token in tokens if token not in stopwords.words('english')]
    # Lemmatize the tokens
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]
    # Join the tokens back into a string
    processed_text = ' '.join(lemmatized_tokens)
    return processed_text
```

```python
# Apply the function df

df = pd.DataFrame({'reviewText': ['This is a great product!', 'It was easy to use and worked well.', 'I would recommend it to anyone.']})

df['reviewText'] = df['reviewText'].apply(preprocess_text)

print(df)
```

```python
# Initialize NLTK sentiment analyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
# Create get_sentiment function

def get_sentiment(text):
    scores = analyzer.polarity_scores(text)
    sentiment = 1 if scores['pos'] > 0 else 0
    return sentiment
```

```python
# Apply get_sentiment function

df['sentiment'] = df['reviewText'].apply(get_sentiment)
df
```

```python
from sklearn.metrics import confusion_matrix

df = pd.DataFrame({'reviewText': ['This is a great product!', 'It was easy to use and worked well.', 'I would recommend it to anyone.'],
                   'sentiment': [1, 1, 1]})

print(confusion_matrix(df['sentiment'], df['sentiment']))
```

```python
from sklearn.metrics import classification_report

df = pd.DataFrame({'reviewText': ['This is a great product!', 'It was easy to use and worked well.', 'I would recommend it to anyone.'],
                   'sentiment': [1, 1, 1]})

print(classification_report(df['sentiment'], df['sentiment']))
```
